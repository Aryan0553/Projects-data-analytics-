```python
#Addiction Level vs Mental Health
df.groupby('addiction_level')[
    ['Mental_Health_Score','mood_score','anxiety_level','sleep_hours']
].mean()
```

```python
#Social Media Usage Impact on Mood, Anxiety & Focus
df.groupby(pd.qcut(df['social_media_time_min'], 4))[
    ['mood_score','anxiety_level','focus_score']
].mean()
```

```python
#Average Mental Health Based on Screen Time Levels
df['screen_time_group'] = pd.cut(
    df['daily_screen_time_min'],
    bins=[0,120,240,360,600,2000],
    labels=['0–2 hrs','2–4 hrs','4–6 hrs','6–10 hrs','10+ hrs']
)

df.groupby('screen_time_group')['Mental_Health_Score'].mean()
```

```python
#Correlation Between Usage & Mental Health
usage_cols = [
    'daily_screen_time_min','social_media_time_min','Avg_Daily_Usage_Hours'
]

mental_cols = [
    'mood_score','anxiety_level','focus_score',
    'digital_wellbeing_score','Mental_Health_Score'
]

df[usage_cols + mental_cols].corr()
```

```python
#top 10 students by social media addiction
df[['Student_ID','addiction_level','daily_screen_time_min','Mental_Health_Score']]\
.sort_values(by='addiction_level', ascending=False).head(10)
```

```python
#conflict vs mental health
df.groupby('Conflicts_Over_Social_Media')[
    ['Mental_Health_Score','anxiety_level','focus_score']
].mean()
```

```python
#Top Countries by Mental Health Score
df.groupby('Country')[['Mental_Health_Score','digital_wellbeing_score','anxiety_level']].mean()\
.sort_values(by='Mental_Health_Score')
```

```python
# Missing Mental Health Score by Platform
df.groupby('Most_Used_Platform')['Mental_Health_Score'].apply(lambda x: x.isna().sum())

# Missing Sleep Hours by Addiction Level
df.groupby('addiction_level')['sleep_hours'].apply(lambda x: x.isna().sum())
```

```python
# Option 1: Fill missing numeric values with mean
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Option 2: Fill missing categorical values with mode
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
missing_matrix = df.isna().astype(int)
missing_correlation = missing_matrix.corr()
missing_correlation['Mental_Health_Score'].sort_values(ascending=False)
```

```python
complete_records = df.dropna().shape[0] / df.shape[0] * 100
print(f"Percentage of complete records: {complete_records:.2f}%")
```
